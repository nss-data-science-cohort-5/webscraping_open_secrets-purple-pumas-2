```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import re
```

```python
#tn_districts = pd.DataFrame(columns = ['Name', 'Party', 'State', 'District', 'Incumbent', 'Winner', 'Raised', 'Spent'])
tn_districts = []

for i in range(1,10):
    url = f'https://www.opensecrets.org/races/summary?cycle=2020&id=TN0{i}&spec=N'
    response = requests.get(url)
    soup = bs(response.text,'html.parser')
    table = soup.findAll('table')[0]
    district = pd.read_html(str(table))[0]
    district['Name'] = district['Candidate'].str.extract(r'([A-Za-z]+\s[A-Za-z]+)')
    district['Party'] = district['Candidate'].str.extract(r'(\((\w{1})\))')[1]
    district['State'] = re.search(r'(id=)(\w{2})', url)[2]
    district['District'] = re.search(r'(id=\w{2})(\d{2})', url)[2]
    district['Incumbent'] = district['Candidate'].str.extract(r'(Incumbent)')
    district['Winner'] = district['Candidate'].str.extract(r'(Winner)')
    district = district.replace(('Incumbent', 'Winner'), ('Yes', 'Yes'))
    district = district.fillna('No')
    district = district[['Name', 'Party', 'State', 'District', 'Incumbent', 'Winner', 'Raised', 'Spent']]
    tn_districts.append(district)

tn_districts = pd.concat(tn_districts).reset_index(drop = True)
```
